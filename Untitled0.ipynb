```python
import math

# Captura de Datos

# Datos Modulo solar - Forma Trapezoidal
mod_solar_b_mayor = 10
mod_solar_b_menor = 4
mod_solar_altura =  4

# Datos Terreno Disponible - Forma Rombo
terreno_diag_mayor = 30
terreno_diag_menor = 16

# C√°lculos
# Areas
area_modulo =((mod_solar_b_mayor + mod_solar_b_menor)* mod_solar_altura)/ 2
area_lote = (terreno_diag_mayor * terreno_diag_menor)/2

#Cantidades
cantidad_modulos = area_lote / area_modulo
cantidad_modulos_completos = math.floor(cantidad_modulos)  # Modulos que caben completos
cantidad_mod_exceso = math.ceil(cantidad_modulos)          # Modulos incluso que se salen

#Area de exceso
area_modulos_exceso = cantidad_mod_exceso * area_modulo   # Los modulos por el area de cada uno
area_completa_exceso = area_modulos_exceso - area_lote    # Area que se excede menos el lote

#Porcentaje de exceso
porc_exceso = (area_completa_exceso / area_lote) * 100

# Visualizacion
print(f'''
En un lote de Dimensiones: {terreno_diag_mayor} m¬≤ x {terreno_diag_menor} m¬≤
Area de cada Modulo: {area_modulo} m¬≤
Cantidad de M√≥dulos con Decimal: {cantidad_modulos}
Cantidad M√≥dulos Completos: {cantidad_modulos_completos}
Cantidad M√≥dulos con Exceso: {cantidad_mod_exceso}

________________________________________
Area Exceso: {area_completa_exceso} m¬≤
Porcentaje de Exceso: {porc_exceso}%

''')
```

```python
import numpy as np
import matplotlib.pyplot as plt

from collections import Counter

### CARGA DE DATOS
ciudades = ['Bogot√°', 'Medell√≠n', 'Cali', 'Barranquilla', 'Cartagena', 'Manizales',
'Pereira', 'Bucaramanga', 'C√∫cuta', 'Ibagu√©']
energias = ['Hidr√°ulica', 'Solar', 'E√≥lica', 'Biomasa', 'Geot√©rmica', 'Carb√≥n',
'Gas Natural', 'Diesel']

### GENERACION DE LAS MUESTRAS ALEATORIAS
tam_muestra = 100
muestra_cuidades = np.random.choice(ciudades, size=tam_muestra )
muestra_energia = np.random.choice(energias, size=tam_muestra)

### CALCULOS
frecuencia_energias = Counter(muestra_energia)

tipos_energia = list(frecuencia_energias.keys())
cantidad_proyectos = list(frecuencia_energias.values())

plt.figure(figsize= (7, 3))           ### ancho y largo
plt.bar(tipos_energia, cantidad_proyectos, color = 'cyan') ## color de las barras
plt.title('Cantidad de Proyectos de Energia Por Tipo')
plt.ylabel('Tipos de Energia')
plt.xlabel('Cantidad de Proyectos')
plt.grid()                                                      ## reglillas

plt.xticks(rotation= 30)   ## rotacion de la letra
plt.tight_layout()
plt.grid(linewidth= 0.4, alpha= 0.5, color='green', linestyle= '--') ### linetyle: lines, alpha: transparencia, color de las lineas.
plt.show()
```

```python
import pandas as pd

### CARGA DE LOS DATOS
url = '/content/Estado_de_la_prestaci√≥n_del_servicio_de_energ√≠a_en_Zonas_No_Interconectadas_20251202.csv'
df_original = pd.read_csv(url)

### FILTRAR EL DATASET POR ALGUNAS COLUMNAS
df = df_original[['A√ëO SERVICIO','DEPARTAMENTO', 'MUNICIPIO', 'ENERG√çA ACTIVA', 'ENERG√çA REACTIVA', 'POTENCIA M√ÅXIMA']]

df['ENERG√çA ACTIVA'] = df['ENERG√çA ACTIVA'].str.replace(',','').astype(float).astype(int)
df['ENERG√çA REACTIVA'] = df['ENERG√çA REACTIVA'].str.replace(',','').astype(float).astype(int)
df['POTENCIA M√ÅXIMA'] = df['POTENCIA M√ÅXIMA'].str.replace(',','').astype(float).astype(int)

lst_cambio = [['√Å','A'], ['√â','E'], ['√ç','I'], ['√ì','O'], ['√ö','U']]

for viejo, nuevo in lst_cambio:
    df['DEPARTAMENTO'] = df['DEPARTAMENTO'].str.replace(viejo,nuevo)
    df['MUNICIPIO'] = df['MUNICIPIO'].str.replace(viejo,nuevo)

condicion = [
    'ARCHIPIELAGO DE SAN ANDRES'
    'ARCHIPIELAGO DE SAN ANDRES Y PROVIDENCIA'
    'ARCHIPIELAGO DE SAN ANDRES, PROVIDENCIA Y SANTA CATALINA'
]

df_cont = df[~df['DEPARTAMENTO'].isin(condicion)]
display(df_cont)
```

```python
### AGRUPAR LOS DATOS PRO DEPARTAMENTOY MUNICIPIO
## SUMANDO los valores de Energia

df_agrupado = df_cont.groupby(['DEPARTAMENTO','MUNICIPIO'])[['ENERG√çA ACTIVA', 'ENERG√çA REACTIVA']].sum()
df_agrupado.to_excel('/content/drive/MyDrive/Estado_de_la_prestacioÃÅn_del_servicio_de_energiÃÅa_en_Zonas_No_Interconectadas_20251202.csvENERGIAS POR DEPTO Y MUNICIPIO.xlsxENERGIAS POR DEPTO Y MUNICIPIO.xlsx')

df_pivote = df_cont.pivot_table(
    index = 'DEPARTAMENTO',
    columns = 'A√ëO SERVICIO',
    values = ['ENERG√çA ACTIVA'],
    aggfunc = 'sum'
)

display(df_pivote)
```

```python
import matplotlib.pyplot as plt

df_pivote.plot(kind='barh', figsize=(8,4
))
plt.title('Energ√≠a Activa por Departamento y A√±o')
plt.xlabel('Energ√≠a en Mw')
plt.tight_layout()
```
